```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
breast_cancer_wisconsin_diagnostic = fetch_ucirepo(id=17) 

# data (as pandas dataframes) 
X = breast_cancer_wisconsin_diagnostic.data.features
Y = breast_cancer_wisconsin_diagnostic.data.targets 
```

<h2 align="center"> <font color='blue'>Exercice 1</font></h2> <a class="anchor" id="1"></a>

```python
plt.figure(figsize=(20, 35), dpi=80)

for i in range(10):
    plt.subplot(5, 2, i+1) #un boxplot pour chacun des 10 attributs
    
    attribute_name = X.columns[i]
    plt.title(attribute_name[:-1])
    
    column1 = np.array(X.iloc[:,i]) #Mean Value
    column2 = np.array(X.iloc[:,i+10]) #Standard Error
    column3 = np.array(X.iloc[:,i+20]) #Largest Value
    
    plt.boxplot([column1, column2, column3])
    plt.xticks(np.arange(4), ['', 'Mean Value', 'Standard Error', 'Largest Value'])
    
plt.show()
```

<h2 align="center"> <font color='blue'>Exercice 2</font></h2> <a class="anchor" id="2"></a>

```python
plt.figure(figsize=(20, 8))
heatmap = sns.heatmap(X.corr(), vmin=-1, vmax=1, annot=True, fmt=".2f")
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12)
```
